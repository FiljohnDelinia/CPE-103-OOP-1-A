```python
#1st Student
def get_valid_score(prompt):
    while True:
        try:
            score = float(input(prompt))
            if 0 <= score <= 100:  #Check if score is between 0 and 100
                return score
            else:
                print("Invalid score. Please enter a score between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

print("Student's Information")
print()
name = input("Enter Student's Name: ")
print()

#PRELIM
PHOA = get_valid_score("Enter Prelim Hands-on Activity grade: ")
PQUIZ = get_valid_score("Enter Prelim Quiz grade: ")
PAS = get_valid_score("Enter Prelim Assignment grade: ")
PCS = PHOA * 0.5 + PQUIZ * 0.3 + PAS * 0.2
print("Prelim CS is: ", PCS)
PEX = get_valid_score("Enter Prelim Exam grade: ")
PG = PEX * 0.5 + PCS * 0.5
print("Prelim Grade is: ", "%.2f" % PG)

print()
#MIDTERM
MHOA = get_valid_score("Enter Midterm Hands-on Activity grade: ")
MQUIZ = get_valid_score("Enter Midterm Quiz grade: ")
MAS = get_valid_score("Enter Midterm Assignment grade: ")
MCS = MHOA * 0.5 + MQUIZ * 0.3 + MAS * 0.2
print("Midterm CS is: ", MCS)
MEX = get_valid_score("Enter Midterm Exam grade: ")
MG = MEX * 0.5 + MCS * 0.5
print("Midterm Grade is: ", "%.2f" % MG)

print()
#FINALS
FHOA = get_valid_score("Enter Finals Hands-on Activity grade: ")
FQUIZ = get_valid_score("Enter Finals Quiz grade: ")
FAS = get_valid_score("Enter Finals Assignment grade: ")
FCS = FHOA * 0.5 + FQUIZ * 0.3 + FAS * 0.2
print("Finals CS is: ", FCS)
FEX = get_valid_score("Enter Finals Exam grade: ")
FG = FEX * 0.5 + FCS * 0.5
print("Finals Grade is: ", "%.2f" % FG)
```

```python
#Student 1
def get_valid_score(prompt):
    while True:
        try:
            score = float(input(prompt))  # Convert input to float
            if 0 <= score <= 100:  # Check if score is between 0 and 100
                return score
            else:
                print("Invalid score. Please enter a score between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def convert_to_ucc_grade(final_grade):
    if 99 <= final_grade <= 100:
        return 1.00
    elif 95 <= final_grade <= 98:
        return 1.25
    elif 90 <= final_grade <= 94:
        return 1.50
    elif 85 <= final_grade <= 89:
        return 1.75
    elif 80 <= final_grade <= 84:
        return 2.00
    elif 75 <= final_grade <= 79:
        return 2.25
    elif 70 <= final_grade <= 74:
        return 2.50
    elif 65 <= final_grade <= 69:
        return 2.75
    elif 60 <= final_grade <= 64:
        return 3.00
    else:
        return 5.00

print("Student's Information")
print()
name = input("Enter Student's Name: ")
print()

# PRELIM
PHOA = get_valid_score("Enter Prelim Hands-on Activity grade: ")
PQUIZ = get_valid_score("Enter Prelim Quiz grade: ")
PAS = get_valid_score("Enter Prelim Assignment grade: ")
PCS = PHOA * 0.5 + PQUIZ * 0.3 + PAS * 0.2
print("Prelim CS is: ", PCS)
PEX = get_valid_score("Enter Prelim Exam grade: ")
PG = PEX * 0.5 + PCS * 0.5
print("Prelim Grade is: ", "%.2f" % PG)

print()
# MIDTERM
MHOA = get_valid_score("Enter Midterm Hands-on Activity grade: ")
MQUIZ = get_valid_score("Enter Midterm Quiz grade: ")
MAS = get_valid_score("Enter Midterm Assignment grade: ")
MCS = MHOA * 0.5 + MQUIZ * 0.3 + MAS * 0.2
print("Midterm CS is: ", MCS)
MEX = get_valid_score("Enter Midterm Exam grade: ")
MG = MEX * 0.5 + MCS * 0.5
print("Midterm Grade is: ", "%.2f" % MG)

print()
# FINALS
FHOA = get_valid_score("Enter Finals Hands-on Activity grade: ")
FQUIZ = get_valid_score("Enter Finals Quiz grade: ")
FAS = get_valid_score("Enter Finals Assignment grade: ")
FCS = FHOA * 0.5 + FQUIZ * 0.3 + FAS * 0.2
print("Finals CS is: ", FCS)
FEX = get_valid_score("Enter Finals Exam grade: ")
FG = FEX * 0.5 + FCS * 0.5
print("Finals Grade is: ", "%.2f" % FG)

# Convert the final grade into the UCC numerical grade
ucc_grade = convert_to_ucc_grade(FG)
print("UCC Numerical Grade: ", ucc_grade)
```
